# Experiment Plots

This notebook takes the processed and stored data and generates figure 4


```python
%pylab inline

rcParamsDefault = plt.rcParams
import pandas as pd

import matplotlib.image as mpimg
import matplotlib as mpl
```

```python
# load the aggregate data

wd = "../results-anonymized/experiment/"
end_of_game = pd.read_csv(wd+"end_of_game_effects_summary.csv", index_col=[0,1])

end_of_game['Effect Upper Error'] = end_of_game['Effect Upper Bound'] - end_of_game['Effect Size']
end_of_game['Effect Lower Error'] = end_of_game['Effect Size'] - end_of_game['Effect Lower Bound']

end_of_game
```

```python
# load raw data for each game (for dots in figure)

raw = pd.read_csv(wd+"end_of_game_measurements.csv")
pairs = raw[raw['network']==0] # only dodec network
cols=['net spoke 5% similarity', 'net spoke 95% similarity', 'net spoke PC1', 
      'survey 5% similarity', 'survey 95% similarity','survey PC1']

# dots represent effect observed in a single game pair
dots = pairs.groupby('id').apply(lambda g: g[g['t']==1][cols].reset_index() - g[g['t']==0][cols].reset_index())
dots
```
