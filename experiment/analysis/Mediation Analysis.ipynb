# Mediation Analysis

```r
#install.packages("lavaan", dependencies=TRUE)
library(lavaan)
```

```r
working_dir = "../results-anonymized/experiment/"
game_factors = read.csv(paste(working_dir, "game_level_factors.csv", sep=""))
game_factors
```

```r
game_outcomes = read.csv(paste(working_dir, "end_of_game_measurements.csv", sep=""))
game_outcomes
```

```r
df = merge(game_factors, game_outcomes)
df
```

```r
#set.seed(1234)
#X <- rnorm(100)
#M <- 0.5*X + rnorm(100)
#Y <- 0.7*M + rnorm(100)
#Data <- data.frame(X = X, Y = Y, M = M)
model <- ' # mediator
             n_shared_edges ~ a*n_triangle

           # effect on PC1
             net.spoke.PC1 ~ PC1.spoke.c*n_triangle  # direct effect
             net.spoke.PC1 ~ PC1.spoke.b*n_shared_edges  # effect of mediator
             net.spoke.PC1 ~ PC1.spoke.n*network  # effect of network structure

             PC1.spoke.indirect := (a*PC1.spoke.b)
             PC1.spoke.total := PC1.spoke.c + (a*PC1.spoke.b)
             PC1.spoke.mediationFrac := (PC1.spoke.indirect / PC1.spoke.total)

             survey.PC1 ~ PC1.survey.c*n_triangle
             survey.PC1 ~ PC1.survey.b*n_shared_edges 
             survey.PC1 ~ PC1.survey.n*network

             PC1.survey.indirect := (a*PC1.survey.b)
             #PC1.survey.total := PC1.survey.c + (a*PC1.survey.b)
             PC1.survey.total := PC1.survey.c + PC1.survey.indirect
             PC1.survey.mediationFrac := (PC1.survey.indirect / PC1.survey.total)
             
             
           # effect on 5% similarity
             net.spoke.5..similarity ~ S5.spoke.c*n_triangle 
             net.spoke.5..similarity ~ S5.spoke.b*n_shared_edges 
             net.spoke.5..similarity ~ S5.spoke.n*network

             S5.spoke.indirect := (a*S5.spoke.b)
             S5.spoke.total := S5.spoke.c + (a*S5.spoke.b)
             S5.spoke.mediationFrac := (S5.spoke.indirect / S5.spoke.total)

             survey.5..similarity ~ S5.survey.c*n_triangle
             survey.5..similarity ~ S5.survey.b*n_shared_edges 
             survey.5..similarity ~ S5.survey.n*network

             S5.survey.indirect := (a*S5.survey.b)
             S5.survey.total := S5.survey.c + (a*S5.survey.b)  
             S5.survey.mediationFrac := (S5.survey.indirect / S5.survey.total)

           # effect on 95% similarity
             net.spoke.95..similarity ~ S95.spoke.c*n_triangle 
             net.spoke.95..similarity ~ S95.spoke.b*n_shared_edges 
             net.spoke.95..similarity ~ S95.spoke.n*network

             S95.spoke.indirect := (a*S95.spoke.b)
             S95.spoke.total := S95.spoke.c + (a*S95.spoke.b)
             S95.spoke.mediationFrac := (S95.spoke.indirect / S95.spoke.total)

             survey.95..similarity ~ S95.survey.c*n_triangle
             survey.95..similarity ~ S95.survey.b*n_shared_edges 
             survey.95..similarity ~ S95.survey.n*network

             S95.survey.indirect := (a*S95.spoke.b)
             S95.survey.total := S95.survey.c + (a*S95.survey.b)
             S95.survey.mediationFrac := (S95.survey.indirect / S95.survey.total)
        

         '
fit <- sem(model, data = df)
summary(fit)
```
